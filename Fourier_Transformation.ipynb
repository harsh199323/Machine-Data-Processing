# Notebook to visualize Fourier Transform and Filtering
## Plot raw signal



```python
import numpy as np
import matplotlib.pyplot as plt

# Define Signal with two components
sample_frequency = 500 # Hz
signal_timewindow = 0.3 # seconds
signal_freq1 = 50       # Hz
signal_freq2 = 120      # Hz

# Generate Signal Data
t = np.linspace(0, signal_timewindow, round(signal_timewindow*sample_frequency))
x = np.sin(signal_freq1 * 2 * np.pi * t) + 0.5 * np.sin(signal_freq2 * 2 * np.pi * t)

# Plot
plt.ylabel("Signal")
plt.xlabel("Time [s]")
plt.plot(t, x)
plt.show()
```

## Compute and visualize Fourier Transform

```python
fft = np.fft.fft(x)
fft
```

```python
f = np.fft.fftfreq(t.shape[-1])*sample_frequency
f
```

```python
# Compute Fast Fourier Transform
fft = np.fft.fft(x)       # complex spectrum
fft_abs = np.abs(fft)     # amplitude spectrum
fft_phase = np.angle(fft) # phase spectrum
f = np.fft.fftfreq(t.shape[-1])*sample_frequency # frequency information

num_samples = x.size
num_half = num_samples//2

# Plot
plt.ylabel("Signal")
plt.xlabel("Time [s]")
plt.plot(t, x)
plt.show()

# Plot Half Amplitude Spectrum
plt.ylabel("Amplitude")
plt.xlabel("Frequency [Hz]")
plt.bar(f[:num_half], fft_abs[:num_half] / num_samples)
plt.show()

# Plot Half Phase Spectrum
plt.ylabel("Amplitude")
plt.xlabel("Frequency [Hz]")
plt.plot(f[:num_half], fft_phase[:num_half] / num_samples)
plt.show()
```

## Apply Windowing Function

```python
window = np.hamming(num_samples)
x_window = x * window

plt.plot(x_window)
plt.show()
```

```python
# Apply Hanning Window
window = np.hamming(num_samples)
x_window = x*window

# Compute Fast Fourier Transform
fft_window = np.fft.fft(x_window)       # complex spectrum
fft_window_abs = np.abs(fft_window)     # amplitude spectrum

# Plot Windowed Data
plt.ylabel("Windowed Signal")
plt.xlabel("Time [s]")
plt.plot(t, x_window)
plt.show()


# Plot Half Amplitude Spectrum
plt.ylabel("Amplitude")
plt.xlabel("Frequency [Hz]")
plt.bar(f[:num_half], fft_abs[:num_half] / num_samples)
plt.show()

# Plot Half Amplitude Spectrum
plt.ylabel("Amplitude")
plt.xlabel("Frequency [Hz]")
plt.bar(f[:num_half], fft_window_abs[:num_half] / num_samples)
plt.show()
```
